```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
movie_streaming = "MoviesOnStreamingPlatforms_updated.csv"
imdb_movie = "tmdb_movies_data.csv"

# Read the mouse data and the study results
movie_streaming = pd.read_csv(movie_streaming)
imdb_movie = pd.read_csv(imdb_movie)

```

```python
movie_streaming["Title"] = movie_streaming["Title"].str.lower()
movie_streaming.head()
```

```python
imdb_movie["original_title"] = imdb_movie["original_title"].str.lower()
imdb_movie.head()
```

```python
# Combine the data into a single dataset
movie_db = pd.merge(movie_streaming, imdb_movie, left_on = ["Title"], right_on = ["original_title"])
movie_db.head()
```

```python
movie_db_nw = movie_db.drop(columns = ["Unnamed: 0","ID", "cast", "homepage", "tagline", "keywords","overview","production_companies","release_date","release_year","budget_adj","revenue_adj","genres","director","runtime","id","imdb_id","original_title","Type"])
movie_db_nw
```

```python
movie_db_nw["Genres"] = movie_db_nw["Genres"].str.split(",", expand = True)[0]
movie_db_nw["Country"] = movie_db_nw["Country"].str.split(",", expand = True)[0]
movie_db_nw["Language"] = movie_db_nw["Language"].str.split(",", expand = True)[0]
movie_db_nw["Directors"] = movie_db_nw["Directors"].str.split(",", expand = True)[0]
```

```python
movie_db_nw
```

```python
movie_db_nw.count()
```

```python
movie_db_nw = movie_db_nw.dropna(how = "any")
movie_db_nw = movie_db_nw.reset_index()
```

```python
movie_db_nw.count()
```

```python
genero = movie_db_nw["Genres"].unique()
genero_df = pd.DataFrame(genero)
genero_df.columns=["genre_name"]
genero_df
```

```python
genre_df2 = genero_df.reset_index()
genre_df2.columns = ["genre_id", "genre_name"]
genre_df2
```

```python
director = movie_db_nw["Directors"].unique()
director_df = pd.DataFrame(director)
director_df.columns=["director_name"]
director_df
```

```python
director_df2 = director_df.reset_index()
director_df2.columns = ["director_id", "director_name"]
director_df2
```

```python
country = movie_db_nw["Country"].unique()
country_df = pd.DataFrame(country)
country_df.columns=["country_name"]
country_df
```

```python
country_df2 = country_df.reset_index()
country_df2.columns = ["country_id", "country_name"]
country_df2
```

```python
movie_rate = movie_db_nw[["IMDb","Rotten Tomatoes", "vote_count", "vote_average", "popularity"]]
movie_rate_df = pd.DataFrame(movie_rate)
movie_rate_df.columns=["imdb_rate", "tomatoe_rate", "imdb_votes_count","imdb_votes", "popularity"]
movie_rate_df
```

```python
movie_rate_df["tomatoe_rate"] = ((movie_rate_df["tomatoe_rate"].str.rstrip('%').astype('float'))/10)
```

```python
movie_rate_df["movie_rate"] = ((movie_rate_df["tomatoe_rate"] + movie_rate_df["imdb_votes"] + movie_rate_df["imdb_rate"] + movie_rate_df["popularity"])/4)
movie_rate_df
```

```python
clasificacion = movie_db_nw["Age"].unique()
clasificacion_df = pd.DataFrame(clasificacion)
clasificacion_df.columns=["clasificacion_name"]
clasificacion_df
```

```python
clasificacion_df2 = clasificacion_df.reset_index()
clasificacion_df2.columns = ["clasificacion_id", "clasificacion_name"]
clasificacion_df2
```

```python
streaming_disp = movie_db_nw[["Netflix","Hulu", "Prime Video", "Disney+"]]
streaming_disp_df = pd.DataFrame(streaming_disp)
streaming_disp_df.columns=["netflix", "hulu", "prime_video","disney"]
streaming_disp_df
```

```python
movie_db_nw.head(1)
```

```python
movies_temp = movie_db_nw[["Title", "Directors", "Genres", "budget", "revenue", "Runtime", "Year","Country", "Age"]]
movies_df = pd.DataFrame(movies_temp)
movies_df.columns=["movie_name", "director_name", "genre", "budget", "revenue","runtime","year", "country", "age"]
movies_df
```

```python
movie_df2 = pd.merge(movies_temp, director_df2, left_on = ["director_name"], right_on= ["director_name"])
movie_df2.head()
```

```python
movie_df2 = pd.merge(movie_df2, country_df2, left_on = ["country"], right_on= ["country_name"])
movie_df2.drop(columns = ["country"])
```

```python
movie_df2 = pd.merge(movie_df2, genre_df2, left_on = ["genre"], right_on= ["genre_name"])
movie_df2.drop(columns = ["genre"])
```

```python
movie_df2 = pd.merge(movie_df2, clasificacion_df2, left_on = ["age"], right_on= ["clasificacion_name"])
movie_df2.drop(columns = ["age"])
```

```python
movie_df2 = movie_df2.drop(columns = ["director_name", "country_name", "genre_name", "clasificacion_name","age", "country", "genre"]) 
```
